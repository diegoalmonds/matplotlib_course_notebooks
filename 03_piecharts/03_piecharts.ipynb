# Bar Charts

```python
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
```

## Basic Chart

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.05, 0] # python portion will stick 0.05 * radius out

plt.pie(slices, labels=labels, 
        explode=explode, 
        shadow=True, # for aesthetic purposes, displays a shadow behind the chart
        startangle=90, # chooses an arbitrary angle to rotate the chart around
        wedgeprops={'edgecolor': 'black'}, # creates a black edge around each portion
        autopct='%1.1f%%') # format string used to display percentages

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
