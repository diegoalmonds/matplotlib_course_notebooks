# Filling Area on Line Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

## Filling Area between Data and Axis

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, # fill_between fills between y1 and y2, in this case py_salaries is y1 and y2 is 0 by default
                 alpha=0.25) # adjusts the transparency of the fill

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
```

## Filling Area between Data and Value

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, overall_median, # overall_median becomes y2
                 alpha=0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
```

## Conditional Fill

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True, alpha=0.25, label='Above Avg') # py_salaries area over the median salary
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries <= overall_median), interpolate=True, color='red', alpha=0.25, label='Below Avg') # py_salaries area under the median salary

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
```
