# Time Series Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mpl_dates
from datetime import datetime, timedelta

plt.style.use('fivethirtyeight')
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

## Formatting X-Axis Labels

```python
# plots the dates and y data with a solid line through all the points
plt.plot_date(dates, y, linestyle='solid')

# automatically formats the labels on the x-axis of the current figure, by default, it rotates them and does other stuff
plt.gcf().autofmt_xdate()

# create a DateFormatter object that formats dates in abbreviated month-day-year format
date_format = mpl_dates.DateFormatter('%b, %d %Y')

# formats the dates on the major ticks (big ticks on the x-axis) as instructed by the DateFormatter object
plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
```

## Plotting Data with Corresponding Times

```python
# this aint a pandas tutorial but this basically makes it so the dates from the csv file are read as dates and not strings and therefore can be sorted
data = pd.read_csv('data.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)
price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
```
